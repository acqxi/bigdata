```
install.packages("datasets")
install.packages("data.table")
library(datasets)
library(data.table)
```

```
car$cyl<-as.factor(car$cyl)#將Month轉為因子變項
car.stat<-data.table(car)[,.(dispMean=mean(disp,na.rm=T),dispSD=sd(disp,na.rm=T)),by=cyl]#計算每月0zone平均與標準差
ggplot(data=car.stat)+#資料airquality.eb
geom_bar(aes(x=cyl,y=dispMean),stat="identity")+#stat="identity"直接畫數字
geom_errorbar(#ymin低點,ymax高點
aes(x=cyl,ymin=dispMean-dispSD,ymax=dispMean+dispSD),width=.1)
```
