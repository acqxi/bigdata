```
install.packages("SportsAnalytics")
```

```
library(SportsAnalytics)
NBA1516<-fetch_NBAPlayerStatistics("15-16")
#導入SportsAnalytics
#並將取得的NBA15、16年資料賦值到變數NBA1516中
```

```
install.packages ("devtools")
devtools::install_github( "cttobin/ggthemr")
library(ggthemr)
```

```
str(NBA1516)
```

```
NBA1516Order<-NBA1516[order(NBA1516$TotalMinutesPlayed,decreasing = T),]
NBA1516Order[1:20,]
#將每個人依據在場時間降冪排序，並印出前20個
```

```
subset(NBA1516,Team=="NYK")
#取得NBA1516子集合，僅包含NYK隊
```

```
NYK1516 = subset(NBA1516,Team=="NYK")
#取得NBA1516子集合，僅包含NYK隊
#並賦值到變數NYK1516中
str(NYK1516)
```

```
NBA1516[grepl("Jose",NBA1516$Name),]
#從NBA1516中取出所有名字中包含JOSE的人
```

```
install.packages("ggplot2")
library(ggplot2)
#導入qqplot2
```

```
ggthemr("earth")
```

```
ggplot(NYK1516,aes(x=Position,y=TotalPoints,color=Name))+geom_point()
#根據位置得到的分數散佈圖，透過顏色將每個人分開
#以下圖表皆僅有NYK
```

```
ggplot()+geom_bar(data=NYK1516,aes(x=Name,y=TotalPoints,fill=TotalMinutesPlayed),binwidth=50,
stat="identity")
#每個人的獲得分數長條圖，顏色說明在場時間
```

```
ggplot(NYK1516,aes(x=TotalMinutesPlayed,y=TotalPoints,color=Position))+geom_line()
#不同位置的人，得分-戰場XY圖
```

```
ggplot(NYK1516,aes(x=TotalMinutesPlayed,y=Assists))+geom_line()+facet_grid(.~Position)
#將不同位置的人的 在場時間-助攻次數 圖 依據位置分開
```

```
bp <- ggplot(NYK1516, aes(x = "", y = TotalPoints, fill = Position    )) +
  geom_bar(width = 1, stat = "identity") 
pie <- bp + coord_polar("y", start = 0)
pie
#依每個位置所貢獻分數比例畫出的圓餅圖
```
