# 不同城市之購買商品數量對比：六都購買量較大

```
dataset %>%
count (CITY) %>%
ggplot(aes(x = fct_reorder(CITY,n), y = n, fill = CITY)) +
geom_col() +
geom_text(aes(label = n), vjust = -0.25) +
labs(title="不同城市之購買商品數量對比",x="CITY",y="QUANTITY")+theme(legend.position ="none")
```

# 各類產品不同月份消費者性別組成：
> 服飾、美容的女性消費者較多，電器、食品男性消費者多，隨新年到來服飾銷售量增加

```
dataset %>%
  count(DATE = substring(DATE, 6, 7),
    MCC_CAT,
    SEX
  ) %>%
  group_by(DATE, MCC_CAT) %>%
  mutate(percent = n / sum(n)) %>%
  ggplot(aes(DATE, n, fill = SEX)) +
  geom_col() +
  facet_wrap(~MCC_CAT) +
  labs(
    x = "購物月份",
    y = "消費比數",
    fill = "性別",
    title = "各類產品不同月份消費者性別組成"
  )
```

# 不同年齡使用不同付款方式之消費者性別組成：
> 付現、玉瓕卡、簽帳金融卡使用人數多，付現人群中男性較多。

```
dataset %>%
  count(AGE = AGE %/% 5,
    CARD_TYPE,
    SEX
  ) %>%
  group_by(AGE, CARD_TYPE) %>%
  mutate(percent = n / sum(n)) %>%
  ggplot(aes(AGE, n, fill = SEX)) +
  geom_col() +
  facet_wrap(~CARD_TYPE) +
  labs(
    x = "年齡",
    y = "付款人次",
    fill = "性別",
    title = "不同年齡使用不同付款方式之消費者性別組成"
  )
```
